```python
import pandas as pd

# Specify the file path
file_path = r'C:\Users\sm291\Documents\ESG_Python_project\Excel_Files\Input_CSV.csv'

# Read the CSV file into a Pandas DataFrame with explicit encoding
df = pd.read_csv(file_path, encoding='latin1')
```

```python
df.head()
```

```python
# Let us check if there are any nulls
# Check for missing values


print(df.isnull().sum())
```

```python
# Handle missing values 
df.dropna(inplace=True)
```

```python
# Making sure they have been taken care of 
print(df.isnull().sum())
```

# The null values have been taken care of

```python

# Display basic information about the dataset
print(df.info())

# Summary statistics for numerical columns
print(df.describe())


# Plot a histogram for ESG ratings from Website 1
plt.figure(figsize=(10, 6))
plt.hist(df['ESG Rating Website 1'].dropna(), bins=20, color='blue', alpha=0.7, label='Website 1')
plt.title('Distribution of ESG Ratings - Website 1')
plt.xlabel('ESG Rating')
plt.ylabel('Frequency')
plt.legend()
plt.show()


```

```python
# Plot a histogram for ESG ratings from Website 2
plt.figure(figsize=(10, 6))
plt.hist(df['ESG Rating Website 2'].dropna(), bins=20, color='green', alpha=0.7, label='Website 2')
plt.title('Distribution of ESG Ratings - Website 2')
plt.xlabel('ESG Rating')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python

# Boxplot to compare ESG ratings from both websites
plt.figure(figsize=(10, 6))
df.boxplot(column=['ESG Rating Website 1', 'ESG Rating Website 2'])
plt.title('Comparison of ESG Ratings between Websites')
plt.ylabel('ESG Rating')
plt.show()
```

```python
# LEt us find the correlation between two website data
correlation = df['ESG Rating Website 1'].corr(df['ESG Rating Website 2'])
print(f"Correlation between Website 1 and Website 2: {correlation}")
```

```python
# Summary statistics for ESG ratings
esg_stats = df[['ESG Rating Website 1', 'ESG Rating Website 2']].describe()
print(esg_stats)
```

```python
import matplotlib.pyplot as plt

# Boxplot for ESG ratings from Website 1
plt.figure(figsize=(10, 6))
df.boxplot(column=['ESG Rating Website 1'])
plt.title('Boxplot of ESG Ratings - Website 1')
plt.ylabel('ESG Rating')
plt.show()

# Boxplot for ESG ratings from Website 2
plt.figure(figsize=(10, 6))
df.boxplot(column=['ESG Rating Website 2'])
plt.title('Boxplot of ESG Ratings - Website 2')
plt.ylabel('ESG Rating')
plt.show()
```
